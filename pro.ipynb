```python
from collections import defaultdict
def scrape1(data):
    
    tab = data.find_all('table', id='MyGridView')[5]
    tabrow = tab.find_all('tr', onmouseover = "this.style.backgroundColor = '#FF8800';")
    num_dict = {} 
    for each in tabrow:
        try:
            temp_dict = defaultdict(dict)
            col = each.find_all('td')
            name = col[0].a.string
            
            for i,val in enumerate(col[1:]):
                #print str(head_list[i].string) + ' ' + str(val.string)
                if i<=2:
                
                    temp_dict[tophead[0]][head_list[i]] = val.string
                  
                elif i<=5:
                    temp_dict[tophead[1]][head_list[i]] = val.string
                    print 'r'
                elif i<=8:
                    temp_dict[tophead[2]][head_list[i]] = val.string
                elif i<=11:
                    temp_dict[tophead[3]][head_list[i]] = val.string
                elif i<=14: 
                    temp_dict[tophead[4]][head_list[i]] = val.string
                print temp_dict
                              
            num_dict[name] = temp_dict 
        except:
            pass
    
    return num_dict
```

```python
playerdict = defaultdict(dict) 

import urllib2
url="http://hoopdata.com/advancedstats.aspx"
page=urllib2.urlopen(url)
soup2 = bs(page.read())

_digits = re.compile('\d')
def contains_digits(d):
    return bool(_digits.search(d))

def fplayerstat(soup4):

    
    for each in soup4.find_all('table')[2].find_all('tr')[1:]:
        #print 
        
        if contains_digits(each.td.text):
            #print each.td.text
            continue
        else:
            name = each.td.text
           
        
        try:
            #print each.td.next_sibling.a.string
            playerdict[name][list_attr[1]] =  each.td.next_sibling.a.string
           
        except:
            pass 
        
        counter = 2
        
        while True: 
            try: 
                playerdict[name][list_attr[counter]] = each.find_all('td')[counter].text
                counter = counter +1 
            except:
                break
        
    return playerdict 
        
    
playerdict = fplayerstat(soup2)  
```
