## Team Data Collection

Here we implement functions to scrape team data from the hoopdata site

```python
"""
Function: get_team_names()
	Get the team names

Parameters:
	None
    
Returns:
	team_data: A list with the abreviated team name
"""
def get_team_names():
    url = 'http://www.hoopdata.com/teamff.aspx?yr=2012&type=pg'
    txt = requests.get(url).text

    #Import data to beautiful soup
    soup = bs(txt)
        
    #Get all tables and accumulate a list of sub_headers
    tables = soup.find_all('table', id='MyGridView')
    sub_headers = [header.a.string for header in tables[5].find_all('th')[1:]]
    
    #Get the relevant table and table rows
    tab = soup.find_all('table', id='MyGridView')[5]
    tabrow = tab.find_all('tr', onmouseover = "this.style.backgroundColor = '#FF8800';")
    
    #Iterate through each row, to get the team names
    
    teams = []
    for row in tabrow:
        #Get all columns in this row
        col = row.find_all('td')
            
        #Make a list of the team names
        teams.append(str(col[0].a.string))
        
    return teams
```

```python

"""
Function: get_team_data()
	Get the team data

Parameters:
	None
    
Returns:
	team_data: A dict containing team data keyed on abbreviated team name
"""
def get_team_data():
    
    #Define category headers
    headers = ['overall','efg','ftr','tor','orr']
            
    #Get list of team names, years
    teams = get_team_names()
    years = range(2007, 2013)
    
    #Initialize the dict we will be returning
    team_data = defaultdict(dict)
    
    #Iteratate over years for which there is data available
    for year in years:
        #Get text from hoopdata page, concating year to be scraped
        url = 'http://www.hoopdata.com/teamff.aspx?yr=' + str(year) + '&type=pg'
        txt = requests.get(url).text

        #Import data to beautiful soup
        soup = bs(txt)
        
        #Get all tables and accumulate a list of sub_headers
        tables = soup.find_all('table', id='MyGridView')
        sub_headers = [header.a.string for header in tables[5].find_all('th')[1:]]
        
        #Get the relevant table and table rows
        tab = soup.find_all('table', id='MyGridView')[5]
        tabrow = tab.find_all('tr', onmouseover = "this.style.backgroundColor = '#FF8800';")
        
        #Iterate through each row, scraping the data
        for row in tabrow:
            temp_dict = defaultdict(dict)
            
            #Get all columns in this row
            col = row.find_all('td')
            
            #Store the team name
            team = str(col[0].a.string)
            
            #Iterate over the columns storing the data in the relevant category
            for i,val in enumerate(col[1:]):
                
                # Comments here follow the follow notation:
                # Category Header - Subheader1, Subheader2, Subheader3
                
                # Overall - OffEff, DefEff, Diff
                if i<=2:
                    temp_dict[headers[0]][sub_headers[i]] = val.string
                    
                # EFG% - Own, Opp, Diff
                elif i<=5:
                    temp_dict[headers[1]][sub_headers[i]] = val.string
                    
                # FTR - Own, Opp, Diff
                elif i<=8:
                    temp_dict[headers[2]][sub_headers[i]] = val.string
                
                # TOR - Own, Opp, Diff
                elif i<=11:
                    temp_dict[headers[3]][sub_headers[i]] = val.string
                # ORR - Own, Opp, Diff
                elif i<=14: 
                    temp_dict[headers[4]][sub_headers[i]] = val.string
                    
            #Account for franchise move
            if team == 'SEA':
                team = 'OKC'
            
            # Add the accumulated data to the finalized dict
            team_data[team][year] = temp_dict
    
    return team_data
```

```python
"""
Function: get_game_data(year)
	Get the result of each game for a given year

Parameters:
	year: The year to scrape game results from
    
Returns:
	game_data: A dict containing a tuple of the date and final score of the game keyed on the teams playing. That is:
    {(team1, team2): (team1_score, team2_score)}
"""
def get_game_data(year):
    
    #Initialize the dict we will be returning
    game_data = defaultdict(list)
    
    #Get text from website, concating year to be scraped
    url = 'http://www.basketball-reference.com/leagues/NBA_' + str(year) + '_games.html#games::none'
    txt = requests.get(url).text

    #Import data to beautiful soup
    soup = bs(txt, 'html.parser') 
    
    #Get desired table and iterable of rows
    table = soup.find('table', id='games')
    rows = table.find_all('tr')[1:]
    
    #Iterate over the rows of the table, getting the name and scores
    for row in rows:
        #Find all data lines in each table row
        lines = row.find_all('td')

        #Get the date of each game by parsing the query url
        query_url = lines[0].a['href'].split('?')[1]
        date_parse = parse_qsl(query_url)
        get_date_info = itemgetter(1)
        
        # Get the date values, cast to integer, and create date in standard notation
        month, day, yr = map(get_date_info, date_parse)
        date = make_date(int(yr), int(month), int(day))
        
        #Get visitor information
        visit =  str(lines[2].text)
        vist_score = int(lines[3].text)
        
        #Get home team info
        home = str(lines[4].text)
        home_score = int(lines[5].text)
        
        #Convert full names to abreviated names for keying
        home_abrv = get_abrv_name(home)
        vist_abrv = get_abrv_name(visit)
        
        #Save the score data
        game_data[(home_abrv, vist_abrv)].append((date, home_score, vist_score))

    return game_data
        
    
```

## Get League Standings

```python
"""
Function: get_league_standings(year)
	Get the league standings for the year requested

Parameters:
	year: The year to collect standings from
    
Returns:
	game_data: A dict containing league wide standings keyed on the abreviated team name. Specifically:
    {abrv_team_name: league_standing}
"""
def get_league_standings(year):
    #Initialize the dict we will be returning
    league_standings = {}
    
    #Get text from website, concating year to be scraped
    url = 'http://www.basketball-reference.com/leagues/NBA_' + str(year) + '_standings.html'
    txt = requests.get(url).text

    #Import data to beautiful soup
    soup = bs(txt, 'html.parser') 
    
    #Get desired table and iterable of rows
    standings_table = soup.find('table', id='expanded-standings')
    rows = standings_table.tbody.find_all('tr')
    
    #Iterate over the rows of the table, getting the name and scores
    for rnk, row in enumerate(rows):
    
        #Get team name and convert to abreviated name
        unabrv_name = str(row.a.text)
        abrv_name = get_abrv_name(unabrv_name)
        
        #Calculate and save rank (adjust for 0-indexing)
        league_standings[abrv_name] = rnk + 1
        
    return league_standings
```

# Modeling

We begin our modeling with a very basic prediction engine. Below I create a preliminary model by examining the average score difference realized when a team ranked n plays a team ranked n+1 (rankings coming from the previous year). We look at data from 2007 forward.

```python
#Define the years to be examined
years = range(2007, 2013)

#Initialize variables
standings = defaultdict(list)
game_data = defaultdict(list)
score_diff = defaultdict(list)

#Iterater over and save the historical data
for year in years:
    #Get game data for given year and the rankings data for the year prio
    standings[year] = get_league_standings(year-1) #Account for year notation difference
    game_data[year] = get_game_data(year)

    #Get the list of team names, create the list of permutations
    teams = standings[year].keys()
    perms = permutations(teams, 2)
    
    #Iterate over each matchup
    for matchup in perms:
        
        #Get the difference in rank between the home team and the visitors
        #This is calculated using ***(VISITOR - HOME)*** that means that if home is higher ranked then rnk_diff > 0
        rnk_diff = standings[year][matchup[1]] - standings[year][matchup[0]]
        
        #Iterate over the games they played and store the score difference (home - visitor)
        for game in game_data[year][matchup]:
            score_diff[rnk_diff].append(game[1] - game[2])
```

```python
avg_diff = defaultdict(list)
diff_std = defaultdict(list)

#Calculate the average point difference and the std. deviation
for rnk_diff in score_diff.keys():
    #Only store data for rank diffences with enough observations to draw statisically significant observations
    req_obvs = pow((1.281551*np.std(score_diff[rnk_diff]))/np.mean(score_diff[rnk_diff]),2)
    num_obvs = len(score_diff[rnk_diff])
    if  num_obvs > req_obvs:
        avg_diff[rnk_diff] = np.mean(score_diff[rnk_diff])
        diff_std[rnk_diff] = np.std(score_diff[rnk_diff])
```

Having calculated the average and standard deviation of score difference, we move forward by first visualizing our findings and then applying them to the most recent 

```python
#Scatter average score difference
plt.scatter(avg_diff.keys(), avg_diff.values())

#Calcuate the best fit line and predict funciton
m, b = np.polyfit(avg_diff.keys(), avg_diff.values(), 1)
predict = lambda x: m*x+b

#plot best fitline
plt.plot(avg_diff.keys(), map(predict, avg_diff.keys()))

#Add plot labels and legend
plt.title('Observed Score Difference vs. Ranking Difference')
plt.ylabel('Score Difference')
plt.xlabel('Standing Difference (Visitor - Home)')
plt.legend(['Best Fit Line', 'Average Score Difference'], loc = 0)

remove_border()
```

We take a moment to interpret our findings. This linear model predicts that each ranking point (from the year prior) is associated with a 0.302 increase in score difference. This finding is inline with our hypothesis that teams ranked higher in year n-1 would outperform lower ranked teams in games played in year n. This model also provides some insight on the value of a home team advantage. In particular, the intercept value of 2.991 represents the predicted difference if a team were to play itself. Holding all other things equal, this value represents the benefit a team recieves from playing at home versus on the road.

The final step before going on to use this model to predict game outcomes is to develop a quantitative understanding of why this approach should provide meaningful insight. Below, we visualize how team standings change from year n to year n+1. Notably the correlation between rankings in one year to rankings the next year is strong (r = 0.6334) meaning that there is a some consistancy to rankings between years. However, the absolute rank change has a mean of 5.4667 and a standard deviation of 5.0049 this shows weakness in our assumption that rankings do not change from one year to the next. However, by virtue of change in rankings being zero-sum, our errors are normally distributed. We take these thoughts into account as we continue building our model.

```python
#Looking at years during which no teams joined or left the league
years = range(2005, 2014)
#Initialize variables
standings = defaultdict(list)
rnk_prev = []
rnk_post = []
rnk_chng = []

#Iterate over and save the historical data
for year in years:
    #Get rankings data
    standings[year] = get_league_standings(year)

#Get team names
teams = standings[year].keys()
    
#Get the change in ranking from year to year
for year in years[1:]: 
    for team in teams:
        #store change in rank
        rnk_prev.append(standings[year-1][team])
        rnk_post.append(standings[year][team])
        rnk_chng.append(standings[year-1][team] - standings[year][team])
```

```python
#Plot how teams change in rank from year to year
plt.scatter(rnk_prev, rnk_post)

#Print metrics
print 'Correlation Coefficient: ' + str(round(pearsonr(rnk_prev, rnk_post)[0],4))
print 'Average Absolute Change: ' + str(round(np.mean(np.abs(rnk_chng)),4))
print 'Std. Dev. of Absolute Rank Change: ' + str(round(np.std(np.abs(rnk_chng)),4))

#Add labels, title
plt.ylabel('Standing in Year n')
plt.xlabel('Standing in Year n-1')
plt.title('Change in League Standing Over 1 Year')
remove_border()


#Plot distribution of rank changes
plt.figure(2)

#Create histogram
plt.hist(rnk_chng, bins=25, normed= True)

#Add labels, title
plt.xlabel('Year over Year Rank Change')
plt.ylabel('Probability Density')
plt.title('Disitribution of Rank Changes')
remove_border()
```

## The Game Network & Model Uncertainty

The previous model was intentionally simplistic and lacked any measure of certainty. Here we add uncertainty to the model, examine results in terms of a distribution rather than an absolutist outcome, and formalize the model as a function.

Furthermore, the previous model relied exclusively on historical data to predict results. In this part of development we go one step further and incorporate networks to update leauge standings. We use the net degree of each node (define by degree_in - degree_out) to construct a measure of team ability. 

```python
"""
Function
--------
game_graph

Parameters
----------
games : dict of lists
    The vote database returned from get_vote_data
    
end_date: datetime.date
    The date to stop analyzing data. Defaults to today's day

Returns
-------
graph : NetworkX Graph object, with the following properties
    1. Each node in the graph is labeled by its abreviated team name
    2. The edges between two nodes are weighted by the number of 
       games the teames have played against each other
    3. Each edge also has a `difference` attribute, which measures the
       net score difference over all games played
"""
def get_game_graph(games, end_date = make_date.today()):
    
    #Initialzie vars
    G = nx.DiGraph()
    
    #Iterate over each matchup
    for matchup, game_data in games.items():
        
        #Iterate over each game in the matchup
        for game in game_data:
            
            #Only add edges of games that are in our examination period
            if game[0] < end_date:
            
                #Created directed edge from loser to winner
                if game[1] > game[2]:
                    
                    #Store the difference, weight, and date
                    if G.has_edge(matchup[0], matchup[1]):
                        G[matchup[0]][matchup[1]]['difference'] += float(game[1] - game[2])
                        G[matchup[0]][matchup[1]]['weight'] += 1.
                        G[matchup[0]][matchup[1]]['date'] = game[0]
                        
                        #Scale the difference by # of games played (this gives the average difference in a win
                        G[matchup[0]][matchup[1]]['difference'] /= G[matchup[0]][matchup[1]]['weight']
                        
                    else:
                        #Create the edge if it doesn't exist
                        G.add_edge(matchup[0], matchup[1], weight=1., difference= float(game[1] - game[2]), date= game[0])
                        
                else:
                    
                    #Store the difference, weight, and date
                    if G.has_edge(matchup[1], matchup[0]):
                        G[matchup[1]][matchup[0]]['difference'] += float(game[2] - game[1])
                        G[matchup[1]][matchup[0]]['weight'] += 1.099832715976 # adjust for the probability of a winning away
                        G[matchup[1]][matchup[0]]['date'] = game[0]
                        
                        #Scale the difference by # of games played (this gives the average difference in a win
                        G[matchup[1]][matchup[0]]['difference'] /= G[matchup[1]][matchup[0]]['weight']
                        
                    else:
                        #Create the edge if it doesn't exist
                        G.add_edge(matchup[1], matchup[0], weight=1.099832715976, difference= float(game[2] - game[1]), date= game[0])
                
    #Create nodes for each team
    for team_name in get_team_names():
        G.add_node(team_name)
        
    return G
```

```python
"""
Function
--------
network_standings

Parameters
----------
year: int
    The year for which rankings are being calculated

Returns
-------
graph_standings: dict
    A dict keyed on the abreviation of the team and the value being the current standings
"""
def network_standings(year, end_date = None):
    #Load game data
    games = get_game_data(year)
    
    #Create the game graph using the appropiate parameters
    if end_date == None:
        game_graph = get_game_graph(games)
    else:
        game_graph = get_game_graph(games, end_date)
        
    #Calculate the net standings of each team
    in_game_deg =game_graph.in_degree(weight= 'weight')
    out_game_deg = game_graph.out_degree(weight = 'weight')
    net_games = dict()
    
    for team in in_game_deg.keys():
        net_games[team] = in_games[team] - out_games[team]

    srted_results = sorted(net_games.iteritems(), key=itemgetter(1))
    
    #Create the dict with rankings
    graph_standings = dict()
    for i, rating in enumerate(srted_results):
        graph_standings[rating[0]] = i + 1
    
    return graph_standings
```

```python
"""
Function
--------
make_ranking_model

Parameters
----------
home_team:
    String of the home team's abbreviated name
    
away_team:
    String of the visiting team's abbreviated name
    
game_data:
    A dict to be used for anyalytis, keyed on year. Each year contains another dict for each matchup.
    This second set of dicts is keyed on the matchup that is: (home_team, away_team)
    
end_date:
    Date to stop analyzing game data. Defaults to today.
    
standing_fn:
    A function to get a rankings dict. Defaults to get_league_standings (which retrieves the previous year's data)

Returns
-------
rnk_diff_distr: A dict keyed on the ranking difference between two teams. Each entry contains a tuple with 3 entries. These are:
                (score_diff_mean, score_diff_std, scores). This is the mean and standard deviation of observed score difference for
                teams differeing in rank by rnk_diff. The final entry is list off all observed scores of that ranking difference.
"""

def make_ranking_model(game_data, end_date =  make_date.today(), standing_fn = get_league_standings):
    
    #Define the years to be examined
    years = game_data.keys()

    #Initialize variables
    standings = defaultdict(list)
    score_diff = defaultdict(list)
    rnk_diff_distr = dict()
    std_devs = list()
    
    #Get the list of team names, create the list of permutations
    teams = get_team_names()
    perms = permutations(teams, 2)
    
    #Iterate over and save the historical data
    for year in years:
        
        #Use the network function for the most recent year using the provided standing fn or the default

        if year == years[-1]:
            #Determine how many arguments we should pass
            if standing_fn == get_league_standings:
                standings[year] = standing_fn(year-1) #Account for year notation difference
            else:
                standings[year] = standing_fn(year, end_date)
        else:
            standings[year] = get_league_standings(year-1) #Account for year notation difference

        #Iterate over each matchup
        for matchup in perms:
            
            #Get the difference in rank between the home team and the visitors
            #This is calculated using ***(VISITOR - HOME)*** that means that if home is higher ranked then rnk_diff > 0
            rnk_diff = standings[year][matchup[1]] - standings[year][matchup[0]]
            
            #Iterate over the games they played and store the score difference (home - visitor)
            for game in game_data[year][matchup]:
                
                #Store score difference if game date is in examination period
                if game[0] < end_date:
                    score_diff[rnk_diff].append(game[1] - game[2])
   
                    
    #Get the average standard deviation, we use this to fill in spaces where there is insufficient data
    for rnk_diff_std, scores_std in score_diff.iteritems():
        
        #Count if min number of observations are made
        if len(scores_std) > 10:
            std_devs.append(np.std(scores_std))
        
    #Save average standard deviation
    std_avg = np.mean(std_devs)
                
    #Calculate the information of the distribution 
    for rnk_diff, scores in score_diff.iteritems():
        
        #If there are not sufficient observations, resort to our previous model to create the mean
        if len(scores) < 5:
            score_mean = predict(rnk_diff)
            score_std =  std_avg
        elif len(scores) < 10:
            score_mean = (predict(rnk_diff)+np.mean(scores))/2.
            score_std = (np.std(scores) + std_avg)/2.
        else:
            score_mean = np.mean(scores)
            score_std = np.std(scores)
          
        #Build dict to be returned
        rnk_diff_distr[rnk_diff] = (score_mean, score_std, scores)
        
    return rnk_diff_distr
```

## Distribution of Points per Game by Team

We examine in this section the distribution of each team's points per game. While this model is fairly naive it allows us to enhance our current prediction methedology.

```python
"""
Function
--------
get_ppg

Parameters
----------
games : dict
    The dict of games of the same format defined in get_game_data()
end_date: DateTime.date
    The date to stop collecitng ppg information. Defaults to today's date

Returns
-------
A tuple of three dicts. The first entry has all score info, the second home, the third away

"""

def get_ppg(games, end_date = make_date.today()):
    
    #create a nested defaultdict
    ppg_all = defaultdict(list)
    ppg_home = defaultdict(list)
    ppg_away = defaultdict(list)
    
    #Iterate over the matchups
    for matchup, game_series in games.iteritems():
        
        #Iterate over the game series
        for game in game_series:
            
            #Only look at games which are before our desired end date
            if game[0] < end_date:
                    
                #Store the points scored by the home team
                ppg_home[matchup[0]].append(game[1])
                ppg_all[matchup[0]].append(game[1])
                
                #Store the points scored by the away team
                ppg_away[matchup[1]].append(game[1])
                ppg_all[matchup[1]].append(game[1])
               
    return (ppg_all, ppg_home, ppg_away)
```

```python
"""
Function
--------
predict_ppg_diff:
    Summarizes the likelihood of point difference outcomes

Parameters
----------
home_team : string
    The abreviated name of the home team

away_team: string
    The abreviated name of the away team
    
games : dict
    The dict of games of the same format defined in get_game_data()

end_date: Datetime.Date
    The date to stop analyzing data. Defaults to today

considerVenue: Bool
    Switch to consider home vs. away bias. Defaults to True

Returns
-------
A tuple with the (average ppg difference, ppg difference std, the probability the home team wins, 
and the distribution of observed ppg differences). Prob calculated using normal.

"""
def predict_ppg_diff(home_team, away_team, games, end_date = make_date.today(), considerVenue= True):

    #Gpoints per game data
    ppg_all, ppg_home, ppg_away = get_ppg(games, end_date = end_date)
    
    #Decide if home advantage should be considered
    if considerVenue:
        #Get the observed mean, std of the difference
        diff_mean = np.mean(ppg_home[home_team]) - np.mean(ppg_away[away_team])
        diff_std = np.sqrt(np.var(ppg_away[home_team]) + np.var(ppg_away[away_team]))
        
        #Ensure list length is equal
        min_len = min(len(ppg_home[home_team]), len(ppg_away[away_team]))
        home_ppg = ppg_home[home_team][:min_len]
        away_ppg = ppg_away[away_team][:min_len]
        
    else:
        #Get the observed mean, std of the difference
        diff_mean = np.mean(ppg_all[home_team]) - np.mean(ppg_all[away_team])
        diff_std = np.sqrt(np.var(ppg_all[home_team]) + np.var(ppg_all[away_team]))
        
        #Ensure list length is equal
        min_len = min(len(ppg_all[home_team]), len(ppg_all[away_team]))
        home_ppg = ppg_all[home_team][:min_len]
        away_ppg = ppg_all[away_team][:min_len]
        
        
    #Collect observed distributions
    ppg_diff = [home - away for home, away in zip(home_ppg, away_ppg)]
        
    #Calculate the probability of a home win
    home_win_prob = 1 - normal_dist(diff_mean, diff_std).cdf(0)
        
    return (diff_mean, diff_std, home_win_prob, ppg_diff)
    
```
